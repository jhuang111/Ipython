```python
def sigmoid(x):
    s=1.0/(1.0+np.exp(-x))
    return s
```

```python
print("sigmoid([0,2])="+str(sigmoid(np.array([0,2]))))
```

```python
#第五步：对参数w与b进行初始化
def initialize_with_zero(dim):
    w=np.zeros((dim,1))
    b=0
    assert(w.shape==(dim,1))
    assert(isinstance(b,float) or isinstance(b,int))
    return w,b
```

```python
dim=2
w,b=initialize_with_zero(dim)
print("w="+str(w))
print("b="+str(b))
```

```python
def propagate(w,b,x,y):
    """
    n=x.shape[0]
    m=x.shape[1]
    
    input:
    w:n*1
    b:scalar
    x:n*m
    y:1*m
    
    output:
    cost:scalar
    dw:n*1
    db:scalar
    """
    #activation
    m=x.shape[1]
    A=sigmoid(np.dot(w.T,x)+b)
    cost=-(np.dot(y,np.log(A).T)\
           +np.dot(1.0-y,np.log(1.0-A).T))/m
    dw=np.dot(x,(A-y).T)/m
    db=np.sum(A-y)/m
    assert(dw.shape==w.shape)
    assert(db.dtype==float)
    cost=np.squeeze(cost)
    assert(cost.shape==())
    grads={"dw":dw,"db":db}
    return grads,cost
```

```python
w, b, x, y = np.array([[1.],[2.]]), 2., \
np.array([[1.,2.,-1.],[3.,4.,-3.2]]), np.array([[1,0,1]])
grads,cost=propagate(w,b,x,y)
print("dw="+str(grads["dw"]))
print("db="+str(grads["db"]))
print("cost="+str(cost))
```

```python
#第六步：对w与b进行更新
def optimize(w,b,x,y,num_iterations,learning_rate,print_cost=True):
    """
    """
    costs=[]
    for i in range(num_iterations):
        grads,cost=propagate(w,b,x,y)
        dw=grads["dw"]
        db=grads["db"]
        w=w-learning_rate*dw
        b=b-learning_rate*db
        if i%100==0:
            costs.append(cost)
        if print_cost and i%100==0:
            print("Cost after iteration %i:%f"%(i,cost))
    params={"w":w,"b":b}
    grads={"dw":dw,"db":db}
    return params,grads,costs
```

```python
params, grads, costs = optimize(w, b, x, y,num_iterations= 100, \
                                learning_rate = 0.009, print_cost = False)

print ("w = " + str(params["w"]))
print ("b = " + str(params["b"]))
print ("dw = " + str(grads["dw"]))
print ("db = " + str(grads["db"]))
```

```python
#第七步：使用训练的w与b进行预测
def predict(w,b,x):
    """
    input:
    w:n*1
    b:scalar
    x:n*m
    output:
    y_prediction:1*m
    """
    m=x.shape[1]
    y_prediction=np.zeros((1,m))
    A=sigmoid(np.dot(w.T,x)+b)
    for i in range(m):
        if A[0,i]<=0.5:
            y_prediction[0,i]=0
        else:
            y_prediction[0,i]=1
    assert(y_prediction.shape==(1,m))
    return y_prediction
```

```python
w = np.array([[0.1124579],[0.23106775]])
b = -0.3
X = np.array([[1.,-1.1,-3.2],[1.2,2.,0.1]])
print ("predictions = " + str(predict(w, b, X)))
```

```python
def model(x_train,y_train,x_test,y_test,num_iterations=2000,\
          learning_rate=0.005,print_cost=False):
    w,b=initialize_with_zero(x_train.shape[0])
    parameters,grads,costs=optimize(w,b,x_train,y_train,num_iterations,\
                                    learning_rate,print_cost)
    w=parameters["w"]
    b=parameters["b"]
    y_prediction_train=predict(w,b,x_train)
    y_prediction_test=predict(w,b,x_test)
    print("train accuray {}%".\
          format(100-np.mean(np.abs(y_prediction_train-y_train))*100))
    print("test accuray {}%".\
          format(100-np.mean(np.abs(y_prediction_test-y_test))*100))
    d = {"costs": costs,
         "Y_prediction_test": y_prediction_test, 
         "Y_prediction_train" : y_prediction_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
```

```python
d=model(train_set_x,train_set_y,test_set_x,test_set_y,\
        num_iterations=2000,learning_rate=0.005,print_cost=True)
```

```python
costs=np.squeeze(d['costs'])
plt.plot(costs)
plt.show()
```

```python
#比较不同的学习率效果
learning_rates=[0.01, 0.001, 0.0001]
models={}
for i in learning_rates:
    print("learning rate is:"+str(i))
    models[str(i)]=model(train_set_x,train_set_y,test_set_x,test_set_y,\
                         num_iterations=1500,learning_rate=i,print_cost=False)
    print('\n'+"-------------------------------------------------------------"+'\n')
for i in learning_rates:
    plt.plot(np.squeeze(models[str(i)]["costs"]), \
             label= str(models[str(i)]["learning_rate"]))

plt.ylabel('cost')
plt.xlabel('iterations')

legend = plt.legend(loc='upper center', shadow=True)
frame = legend.get_frame()
frame.set_facecolor('0.90')
plt.show()
```
